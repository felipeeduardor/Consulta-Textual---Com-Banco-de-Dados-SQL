```python
import vanna
from vanna.remote import VannaDefault
```

```python
# Longa Vanna
# o vanna oferecer uma função para obter uma chave de API voce recebera um codigo enviado para o seu e-mail. Voce pode salvar sua chave de API para o usso futuro .

api_key = vanna.get_api_key('felipeeduardor88@gmail.com')
```

# Base de dados

```python
vanna_model_nome = 'chinook'
vn = VannaDefault(model= vanna_model_nome ,api_key =api_key)
```

# **Conectando a base de dados**

```python
vn.connect_to_sqlite('https://github.com/lerocha/chinook-database/raw/master/ChinookDatabase/DataSources/Chinook_Sqlite.sqlite')
```

```python
#quais são os TOP 5 artistas mais vendidos
vn.ask('qual musica mais tocada')
```

```python
# aplicação WEB com Flask Python
from vanna.flask import VannaFlaskApp
app = VannaFlaskApp(vn)
app.run()
```
